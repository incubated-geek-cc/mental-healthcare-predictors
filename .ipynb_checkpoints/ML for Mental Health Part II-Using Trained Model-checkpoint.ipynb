````python
import joblib
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
````

````python
input_data=pd.read_csv("input/test_model.csv")
input_data.head()
````

````python
scaler = MinMaxScaler()
input_data["age"] = scaler.fit_transform(input_data[["age"]])
input_data.head()
````

````python
# Load model from file
classifer = joblib.load("output/model.pkl")

predictions = classifer.predict(input_data)
# Index is tagged to each role as an identifier
results = pd.DataFrame({
    "Index": input_data.index, 
    "Treatment": predictions
})
# Save to file
# results.to_csv("output/results.csv", index=False)
results.head()
````
